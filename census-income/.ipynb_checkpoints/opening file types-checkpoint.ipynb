## Opening and exploring different data types
This project comes with different file formats from the [archive](https://archive.ics.uci.edu/ml/datasets/Adult) such as `.names`, `.data`, `.test`, and `.file`. Here are the ways on how to open each file format.

* [x] DATA File
* [x] NAMES File 
* [x] TEST File

### Open DATA file
For `(.data)`.

---

```python
import pandas as pd

# Set the file path
file_path = 'files/UCI/adult.data'

# Open the file
data = pd.read_csv(file_path)
data.head()
```

### Open NAMES file
For `(.names)`.

---

```python
file_path = 'files/UCI/adult.names'

with open(file_path, 'r') as f:
    data = f.read()
    
print(data)
```

### Open TEST file
For `(.test)`.

---

```python
import pandas as pd

# Set the file path
test_path = 'files/UCI/adult.test'

# Open the file
test_data = pd.read_csv(test_path, skiprows=1)
test_data.head()
```

**Issue**: Column names are not indicated in the `.data` and `.test` files
